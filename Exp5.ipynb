```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'Competitions': ['Archery', 'Archery', 'Archery', 'Archery', 'Archery'],
    'Total': [7, 2, 2, 1, 1]
}
df = pd.DataFrame(data)

# Group by Competitions and sum the total medals
grouped_df = df.groupby('Competitions').sum().reset_index()

# Plot
plt.figure(figsize=(10, 6))
plt.plot(grouped_df['Competitions'], grouped_df['Total'], marker='o')
plt.title('Total Medals by Competition')
plt.xlabel('Competition')
plt.ylabel('Total Medals')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(grouped_df['Competitions'], grouped_df['Total'], marker='o')

# Adding labels
for i, value in enumerate(grouped_df['Total']):
    plt.text(i, value + 0.2, str(value), ha='center')

plt.title('Total Medals by Competition')
plt.xlabel('Competition')
plt.ylabel('Total Medals')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(grouped_df['Competitions'], grouped_df['Total'], marker='s', linestyle='-', color='r')
plt.title('Total Medals by Competition')
plt.xlabel('Competition')
plt.ylabel('Total Medals')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(x='Competitions', data=df)
plt.title('Count of Each Competition')
plt.xlabel('Competition')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(grouped_df['Competitions'], grouped_df['Total'], marker='o', label='Total Medals')

# Adding legend
plt.legend()

plt.title('Total Medals by Competition')
plt.xlabel('Competition')
plt.ylabel('Total Medals')
plt.grid(True)
plt.show()
```
